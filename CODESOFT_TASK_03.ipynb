```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
x= df.drop(columns=['species'],axis=1)
y=df['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

## Predicting Methods

### Logistic Regression

```python
#creating model set prdection process or binary cllassification process
# regression is to
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
ypred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test, ypred)*100
print('Accuracy = ', accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
m1=RandomForestClassifier()
m2=DecisionTreeClassifier()
```

### Decision Tree Classifier


```python
m2.fit(x_train,y_train)
```

```python
y1pred=m2.predict(x_test)
```

```python
accuracy=accuracy_score(y_test, y1pred)*100
print('Accuracy = ', accuracy)
```

### RandomForestClassifier

```python
m1.fit(x_train,y_train)
```

```python
y2pred=m1.predict(x_test)
accuracy=accuracy_score(y_test, y2pred)*100
print('Accuracy = ', accuracy)
```
