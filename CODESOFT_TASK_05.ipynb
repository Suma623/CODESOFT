# Data Preparation

#### Handling Imbalance

```python
df_fraud = df[df['Class'] == 1]
df_non_fraud = df[df['Class'] == 0]
df_non_fraud_sampled = df_non_fraud.sample(df_fraud.shape[0],replace=False,random_state=42)
df_balanced = pd.concat([df_fraud, df_non_fraud_sampled], axis=0).sample(frac=1,replace=False,random_state=42).reset_index().drop('index',axis=1)
df_balanced
```

### Checking the balance of the dataset

```python
fraud_count = df['Class'].value_counts()
fraud_rate = 100 * fraud_count / df.shape[0]
fraud_data = pd.concat([fraud_count,fraud_rate], axis=1).reset_index()
fraud_data.columns = ['Class','Count','Percentage']
fraud_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_balanced.drop('Class',axis=1),df_balanced['Class'],test_size=0.2,random_state=42)
print(f'''x_train: {x_train.shape}
y_train: {y_train.shape}
x_test: {x_test.shape}
y_test: {y_test.shape}''')
```

### Fitting a Random Forest Model

```python
# logistic model pipe
randomForestModel = Pipeline([('scaler',StandardScaler()),('classifier',RandomForestClassifier())])
randomForestModel.fit(x_train,y_train)
```

## Prediction

```python
y_pred = randomForestModel.predict(x_test)
y_pred
```

### Classification Report

```python
cr=classification_report(y_test,y_pred)
print(cr)
```
